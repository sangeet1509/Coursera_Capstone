# Scraping the web page and Building a Data Frame For TORONTO City

```python
import urllib.request, urllib.parse , urllib.error
from bs4 import BeautifulSoup
```

```python
# read this url "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
html = urllib.request.urlopen(url).read() 
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
print("Soup is:", soup.prettify())
```

```python
soup.title
```

```python
soup.title.string
```

```python
#Finding all tables from the HTML file
all_tables=soup.find_all("table")
all_tables
```

```python
#Finding all table whose class is 'wikitable' from the HTML file
right_table=soup.find('table', class_='wikitable')
right_table
```

```python
#Parsing the html file to get the required data to create the table
A=[]
B=[]
C=[]

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells) != 0:
        #cells[0] = cells[0].find(text=True).strip()
        #cells[0].Replace("\n","")
        A.append(cells[0].find(text=True).strip())
        B.append(cells[1].find(text=True).strip())
        C.append(cells[2].find(text=True).strip())
    
    
```

```python
#print("List A is :", A)
#print("List B is :", B)
#print("List C is :", C)
```

```python
#Creating a Pandas DataFrame
import pandas as pd
df=pd.DataFrame(A,columns=['Postal Code'])
df['Borough']=B
df['Neighborhood']=C
df
```

```python
#Dropping the rows which contain "Not assigned" from Borough column
df.drop(df[df['Borough'] == 'Not assigned'].index, inplace = True)
df
```

```python
df.set_index('Postal Code', inplace=True)
df
```

```python
#Replacing '/' in Neighborhood to ','
df['Neighborhood'] = df['Neighborhood'].replace('/', ',', regex=True)
df
```

```python
#Resetting Index
df1 = df.reset_index()
df1
```

```python
df1.shape
```

# Finding the latitude and longitude of each Postal code

```python
#Reading the csv file containing Postal Code latitude and longitude
df_cord = pd.read_csv("Geospatial_Coordinates.csv")
df_cord.head()
```

```python
#Merging the two dataframe to include Longitude and Latitude
neighborhoods = pd.merge(df1, df_cord, how ='inner', on ='Postal Code')
neighborhoods
```

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
#Creating a function to repeat the same process to all the neighborhoods in Downtown Toronto
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

## Analyze each neighborhood

```python
# one hot encoding
dt_toronto_onehot = pd.get_dummies(dt_toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
dt_toronto_onehot['Neighborhoods'] = dt_toronto_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [dt_toronto_onehot.columns[-1]] + list(dt_toronto_onehot.columns[:-1])
dt_toronto_onehot = dt_toronto_onehot[fixed_columns]

dt_toronto_onehot.head()
```

```python
#Examining the size of the new dataframe
dt_toronto_onehot.shape
```

```python
#Grouping rows by neighborhood and by taking the mean of the frequency of occurrence of each category
dt_toronto_grouped = dt_toronto_onehot.groupby('Neighborhoods').mean().reset_index()
dt_toronto_grouped
```

```python
dt_toronto_grouped.shape
```

```python
#Print each neighborhood along with the top 5 most common venues/categories
num_top_venues = 5

for hood in dt_toronto_grouped['Neighborhoods']:
    print("----"+hood+"----")
    temp = dt_toronto_grouped[dt_toronto_grouped['Neighborhoods'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
# From the above table we can see the frequency of each category/venue so we can say the %tage of that category in that
#particular area. For eg: In Berczy Park neighborhood ,the percentage to get a cofeeshop is 9% and so on...
```

```python
#Now Putting the table into a pandas dataframe
#Before that writing a function to sort the values in descending order
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
#Creating New Dataframe and displaying top 10 venues for each neighborhood
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = dt_toronto_grouped['Neighborhoods']

for ind in np.arange(dt_toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(dt_toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

## Clustering Neighborhoods

```python
# set number of clusters
kclusters = 5

dt_toronto_grouped_clustering = dt_toronto_grouped.drop('Neighborhoods', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(dt_toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
#create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

dt_toronto_merged = dt_toronto_data

# merge dt_toronto_grouped with dt_toronto_data to add latitude/longitude for each neighborhood
dt_toronto_merged = dt_toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

dt_toronto_merged.head() # check the last columns!
```

```python
#Finally lets Visualize the resulting clusters
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = manhattan_grouped['Neighborhood']

for ind in np.arange(manhattan_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(manhattan_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```
